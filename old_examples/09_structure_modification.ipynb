# Rotation

```python
from atomrdf import KnowledgeGraph
import atomrdf.build as build
import atomrdf.transform as transform 
```

```python
kg = KnowledgeGraph()
```

```python
s = build.bulk('Al', repetitions=(1,1,1), graph=kg)
```

```python
s.show.all()
```

```python
rotation_vectors = [[ 1, 1, 0],
                [-1, 1, 0],
                [ 0, 0, 1]]
```

```python
srot = transform.rotate(s, rotation_vectors)
```

```python
srot.show.all()
```

```python
kg.visualise( workflow_view=True, layout='dot')
```

# Shear

For shearing a shear vector, plane, and distance needs to be specified

```python
s = build.bulk('Al', repetitions=(3,3,3), graph=kg)
```

```python
s.show.all()
```

```python
plane = [1, 0, 0]
distance = s.box[0][0]/2
distance
```

```python
tsys = transform.translate(s, [0, 2, 0], plane=plane, distance=distance)
```

```python
tsys.show.all()
```

Note that for shear operation, the plane and the shear vector should be perpendicular to each other.

# Translation

Translation operation can be performed in two ways: you can either translate all atoms, or select a plane and then translate atoms above and beyond it, which is equivalent to shearing a system. The second case:

```python
s = build.bulk('Al', repetitions=(3,3,3), graph=kg)
```

```python
s.show.all()
```

We will select half of the system along the [1,0,0] plane

```python
plane = [1, 0, 0]
distance = s.box[0][0]/2
distance
```

```python
tsys = transform.translate(s, [0, 2.5, 0], plane=plane, distance=distance)
```

```python
tsys.show.all()
```

```python
kg.visualise( workflow_view=True, layout='dot')
```
