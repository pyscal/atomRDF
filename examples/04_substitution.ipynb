```python
from atomrdf import KnowledgeGraph
import atomrdf.build as build
```

```python
kg = KnowledgeGraph()
```

### Substitutional defects

```python
struct_Al = build.bulk('Al', graph=kg)
```

```python
struct_Al = build.defect.substitutional(struct_Al,'Mg', indices=[0])
```

```python
struct_Al.show.all()
```
