## VASP calculation

Read in an existing calculation to pyiron. Can also be a calculation done within pyiron

```python
from atomrdf import KnowledgeGraph, System
from pyiron_atomistics import Project
```

```python
pr = Project('wfy1')
```

Import the finished calculation; this can be ignored if the calculation is performed with pyiron|

```python
pr.import_single_calculation('fccSi', copy_raw_files=True)
```

```python
pr.job_table()
```

```python
job = pr.load('fccSi')
```

```python
kg = KnowledgeGraph()
```

```python
kg.add_workflow(job, workflow_environment='pyiron')
```

```python
kg.visualise(workflow_view=True, layout='dot')
```
