# Creating dislocations

```python
from atomrdf import KnowledgeGraph
import atomrdf.build as build
import numpy as np
```

Define the slip system

```python
slip_direction = np.array([1, 0, -1])
slip_plane = np.array([1, 1, 1])

slip_system = [slip_direction, slip_plane]
```

Define burgers vector

Can be defined in multiple ways, for example, as [0.5, 0, -0.5], or simply as 0.5, in which case the slip_direction is multiplied with 0.5

```python
burgers_vector = 0.5
```

Finally dislocation line

```python
dislocation_line = np.array([1, 0, -1])
```

Elastic constants

```python
elastic_constant_dict = {'C11': 169, 'C12': 122, 'C44': 75.4}
```

Create system

```python
kg = KnowledgeGraph()
```

```python
sys = build.defect.dislocation(slip_system,
                                        dislocation_line,
                                        elastic_constant_dict,
                                        burgers_vector=burgers_vector,
                                        element='Cu',
                                           dislocation_type='monopole',
                                        graph=kg,
                                        )
```

```python
sys.show.all()
```

```python
kg.visualise()
```

Query for results

```python
res = kg.query_sample(kg.ontology.terms.ldo.ScrewDislocation)
res
```
