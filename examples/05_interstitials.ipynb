```python
from atomrdf import KnowledgeGraph
import atomrdf.build as build
import numpy as np
```

```python
kg = KnowledgeGraph()
```

```python
sys = build.bulk('Fe', graph=kg)
```

```python
updated_sys = build.defect.interstitial(sys, ['H', 'C'], void_type='tetrahedral')
```

```python
updated_sys.show.all()
```
