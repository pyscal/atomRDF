```python
import numpy as np
from atomrdf import KnowledgeGraph
import atomrdf.build as build
```

## FCC

```python
slip_plane = np.array([1,1,1])
```

```python
kg = KnowledgeGraph()
```

```python
sf = build.defect.stacking_fault(slip_plane, 
    0.5,
    element="Cu", 
    repetitions=(1,1,1),
    vacuum=0.0,
    graph=kg)
```

```python
sf.show.all()
```

## HCP

```python
slip_plane = np.array([0,0,0,1])
slip_direction = np.array([2,-1,-1,0])
```

```python
kg = KnowledgeGraph()
```

```python
sf = build.defect.stacking_fault(slip_plane, 
    0.25,
    slip_direction_a=slip_direction,
    element="Mg", 
    repetitions=(2,2,2),
    vacuum=1.0,
    graph=kg)
```

```python
sf.show.all()
```
